```python
import lane_detection
import cv2, math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Test with singular frame first
img = cv2.cvtColor(cv2.imread("./test/rov_pool.jpg"), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
lines = lane_detection.detect_lines(img)
lane_detection.draw_lines(img, lines)
plt.imshow(img)
```

```python
# Test detect_lines first
cap = cv2.VideoCapture("./test/auv.mkv")
success = cap.grab()
frames = []
i = 0
while success:
    i += 1
    if i % 50 == 0:
        _, frame = cap.retrieve()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        lines = lane_detection.detect_lines(frame)
        lane_detection.draw_lines(frame, lines)
        frames.append(frame)
    success = cap.grab()
width, height, _ = frames[0].shape
plt.imshow(frames[1])
print(width, height)
video = cv2.VideoWriter("output.avi", cv2.VideoWriter_fourcc(*"DIVX"), 15, (width, height))
for frame in frames:
    video.write(frame)
video.release()
```
